# 1. Load the trip data from 2013 to 2023

```python
# Define the base directory containing the trip folders
base_dir = "data"
```

```python
# Initialize an empty list to store DataFrames
data_frames = []
```

```python
# Define a mapping of inconsistent column names to standard column names
column_mapping = {
    "Start Time": "starttime",
    "started_at": "starttime",
    "Stop Time": "stoptime",
    "ended_at": "stoptime",
    "Start Station ID": "start station id",
    "start_station_id": "start station id",
    "Start Station Name": "start station name",
    "start_station_name": "start station name",
    # "start station latitude": "start station latitude",
    "start_lat": "start station latitude",
    # "start station longitude": "start station longitude",
    "start_lng": "start station longitude",
    # "end station id": "end station id",
    "end_station_id": "end station id",
    # "end station name": "end station name",
    "end_station_name": "end station name",
    # "end station latitude": "end station latitude",
    "end_lat": "end station latitude",
    # "end station longitude": "end station longitude",
    "end_lng": "end station longitude",
    "Bike ID": "bikeid",
    "ride_id": "bikeid",
}
```

```python
# Walk through the directory structure
for root, dirs, files in os.walk(base_dir):
    print(root,dirs,files)
    for file in files:
        # Check if the file is a CSV
        if file.endswith(".csv"):
            file_path = os.path.join(root, file)
            # Load the CSV into a DataFrame and append to the list
            try:
                df = pd.read_csv(file_path)
                
                # Standardize column names
                df.rename(columns=column_mapping, inplace=True)

                # Optionally ensure all necessary columns are present
                required_columns = list(column_mapping.values())
                df = df.reindex(columns=required_columns)

                data_frames.append(df)
            except Exception as e:
                print(f"Could not load {file_path}: {e}")
```

```python
# Concatenate all DataFrames into a single DataFrame
if data_frames:
    combined_data = pd.concat(data_frames, ignore_index=True)
    print(f"Loaded {len(data_frames)} CSV files into a single DataFrame.")
else:
    print("No CSV files were found in the specified directory.")
```

```python
# def custom_parser(date_str):
#     from datetime import datetime
#     try:
#         return datetime.strptime(date_str, '%m/%d/%Y %H:%M:%S')  # MM/DD/YYYY
#     except ValueError:
#         try:
#             return datetime.strptime(date_str, '%Y-%m-%d %H:%M:%S')  # YYYY-MM-DD
#         except ValueError:
#             return pd.NaT  # Return NaT if parsing fails
# def custom_parser(date_str):
#     from datetime import datetime
#     try:
#         # Format: MM/DD/YYYY HH:MM:SS
#         return datetime.strptime(date_str, '%m/%d/%Y %H:%M:%S')
#     except ValueError:
#         try:
#             # Format: YYYY-MM-DD HH:MM:SS
#             return datetime.strptime(date_str, '%Y-%m-%d %H:%M:%S')
#         except ValueError:
#             try:
#                 # Format: YYYY-MM-DD HH:MM:SS.sssss (microseconds)
#                 # Strip microseconds before parsing
#                 trimmed_date_str = date_str.split('.')[0]
#                 return datetime.strptime(trimmed_date_str, '%Y-%m-%d %H:%M:%S')
#             except ValueError:
#                 return pd.NaT  # Return NaT if parsing fails

def custom_parser(date_str):
    from datetime import datetime
    # Check if the input is a string, otherwise handle it
    if isinstance(date_str, str):  # Ensure the input is a string
        try:
            # Format: MM/DD/YYYY HH:MM:SS
            return datetime.strptime(date_str, '%m/%d/%Y %H:%M:%S')
        except ValueError:
            try:
                # Format: YYYY-MM-DD HH:MM:SS
                return datetime.strptime(date_str, '%Y-%m-%d %H:%M:%S')
            except ValueError:
                try:
                    # Format: YYYY-MM-DD HH:MM:SS.sssss (microseconds)
                    # Strip microseconds before parsing
                    trimmed_date_str = date_str.split('.')[0]
                    return datetime.strptime(trimmed_date_str, '%Y-%m-%d %H:%M:%S')
                except ValueError:
                    return pd.NaT  # Return NaT if parsing fails
    elif isinstance(date_str, (int, float)):  # If the input is an integer or float
        try:
            # Convert Unix timestamp to datetime
            return datetime.fromtimestamp(date_str)
        except (OSError, OverflowError):  # Handle invalid or out-of-range timestamps
            return pd.NaT
    return pd.NaT
```

# 3. Retrain and test your model by excluding the data collected during the COVID-19 period. Report the model's prediction performance again and compare it to the results obtained using the complete dataset

```python
# Exclude COVID-19 period
train_data_no_covid = train_data[
    (train_data.index < '2020-03-01') | (train_data.index > '2021-06-30')
]
test_data_no_covid = daily_data[daily_data.index.year == 2023]
```

```python
# Define features and target variable
features = ['day_of_week', 'month', 'is_weekend','year'] # 'trip_count_lag1', 'trip_count_lag7', 'trip_count_rolling7'
X_train_no_covid = train_data_no_covid[features]
y_train_no_covid = train_data_no_covid['trip_count']
X_test_no_covid = test_data_no_covid[features]
y_test_no_covid = test_data_no_covid['trip_count']
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Initialize and train the model
model_no_covid = RandomForestRegressor(n_estimators=100, random_state=42)
model_no_covid.fit(X_train_no_covid, y_train_no_covid)

# Predict on the test set
predictions_no_covid = model_no_covid.predict(X_test_no_covid)
```

```python
from tslearn.clustering import TimeSeriesKMeans
```

```python
X = hourly_trips.values[:, :, np.newaxis] 

# Perform time series KMeans clustering
n_clusters = 5  # Example, you can adjust the number of clusters
model = TimeSeriesKMeans(n_clusters=n_clusters, metric="dtw", verbose=True)
y_pred = model.fit_predict(X)

# Add the predicted cluster to the dataframe
hourly_trips['cluster'] = y_pred
```

```python
import matplotlib.pyplot as plt

# Plot the average demand for each cluster
plt.figure(figsize=(10, 6))

for cluster_id in range(n_clusters):
    # Get the stations in this cluster
    cluster_stations = hourly_trips[hourly_trips['cluster'] == cluster_id]
    
    # Plot the average trip count for this cluster
    avg_demand = cluster_stations.drop('cluster', axis=1).mean(axis=0)
    plt.plot(avg_demand, label=f'Cluster {cluster_id}')

plt.xlabel('Hour of Day')
plt.ylabel('Average Number of Trips')
plt.title('Average Trip Demand by Hour of Day for Each Cluster')
plt.legend()
plt.show()
```
